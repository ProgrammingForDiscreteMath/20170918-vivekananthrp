## Getting the spanning tree from a traversal

Modify the code for graph traversal below to return the resulting spanning tree, rather than just the traversal.

### Hint:
You need to maintain a pointer to where each node in ``can_visit`` "came from".

```python
from collections import deque
from copy import deepcopy
def traverse(G, root, algorithm):
    r"""
    Return a traversal of ``G`` starting with root node ``r`` using DFS or BFS.
    
    INPUT:
    
    ``G`` - a graph or directed graph
    ``r`` - a node of ``G``
    ``algorithm`` - 'DFS' or 'BFS'
    
    Computes and stores a dictionary ``spanning_dict`` of can_visit nodes for each node while traversing.
    
    Using this dictionary edge-list is prepared for the spanning tree for BFS or DFS.
    
    """
    traversal = []
    
    can_visit = deque([root])
    
    spanning_dict = {}
    
    spanning = []
      
    while can_visit: # this runs so long as can_visit is non-empty
        if algorithm == 'DFS':
            n = can_visit.pop()
        elif algorithm == 'BFS':
            n = can_visit.popleft()
        else:
            raise ValueError('algorithm should be DFS or BFS')
        traversal.append(n)
        
        temp2 = [v for v in G.neighbors(n) if not (v in can_visit or v in traversal)]
        
        edge_list = []
        
        spanning_dict[n] = temp2

        if algorithm == 'BFS':
            for k in spanning_dict.keys():
                if (len(spanning_dict[k])>0):
                    edge_list.extend([(k,i) for i in spanning_dict[k]])
                    
        can_visit.extend(temp2)
    
    if algorithm == 'DFS':
        x = deepcopy(root)
        y = spanning_dict[x].pop()
        edge_list.append((x,y))
        
        while True:
            if len(spanning_dict[y]) > 0:
                x = deepcopy(y)
                y = spanning_dict[x].pop()
                edge_list.append((x,y))
            else:
                temp = deepcopy(edge_list)
                temp.pop()
                
                while True:
                    x,y = temp.pop()
                    if len(spanning_dict[x]) > 0:
                        x = deepcopy(x)
                        y = spanning_dict[x].pop()
                        edge_list.append((x,y))
                        break
            
            if sum([len(i) for i in spanning_dict.values()])==0:
                break
                
    return traversal, edge_list
```
